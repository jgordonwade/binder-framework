```python
#Considering only NTAs with at least 5 listed businesses, what is the average total savings 
# and the total jobs created for each NTA?
NTA_ColName='Neighborhood Tabulation Area (NTA) (2020)'
NTA_list = EnergySavingsFrame[NTA_ColName].dropna().unique().tolist()
print(' There are',len(NTA_list), "distinct NTA's")


NTA_dict = {}
for nta in NTA_list:
    ntarows = EnergySavingsFrame.index[EnergySavingsFrame[NTA_ColName] == nta].tolist()
    l = len(ntarows)
    if l>=5:
        NTA_dict[nta]=ntarows
    
print('Got the NTA_dict, which has a row-list for each of the', \
       len(NTA_dict), 'distinct ntas with at least five companies')    


AvgTotalSavings_dict = {}
TotalJobsCreated_dict = {}
for nta, ntarows in NTA_dict.items():
    AvgTotalSavings_dict[nta] = round(EnergySavingsFrame.loc[ntarows,'Total Savings'].sum(),2)
    TotalJobsCreated_dict[nta] = int(EnergySavingsFrame.loc[ntarows,'Job created'].sum())
    
print("Now got two more dict's, one with (nta,avg savings) pairs, \
and one with (nta, jobs created) pairs")

print("\n Now I'm going to write the CSV file per the instructions")
    
filename = 'NYC_EnergySavings.csv'
with open(filename, 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['NTA', 'AVg Savings', 'Jobs Created'])
    for nta in NTA_dict:
        writer.writerow([nta, AvgTotalSavings_dict[nta], TotalJobsCreated_dict[nta]])
        
print('OK Now done! Have a look')
!ls 
```
